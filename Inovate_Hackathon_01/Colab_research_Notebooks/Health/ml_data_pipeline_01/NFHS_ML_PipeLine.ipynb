#cleaning this dataset

```python
df.columns = df.columns.str.strip().str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
import numpy as np

# Replace problematic values
df.replace({'\*': np.nan, '\(': '', '\)': ''}, regex=True, inplace=True)
```

```python
for col in df.columns:
    if col not in ['district', 'state']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['district'] = df['district'].str.strip().str.lower()
df['state'] = df['state'].str.strip().str.lower()
```

```python
df.info()
```

```python
df.describe()
```

```python
for col in df.select_dtypes(include=['float64']).columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    df[col] = df[col].clip(lower, upper)
```

#feature Engineering

```python
df['overall_child_health_risk_score'] = (
    0.25 * df['child_stunting_rate'] +
    0.15 * df['child_wasting_rate'] +
    0.15 * df['child_underweight_rate'] +
    0.15 * df['child_anemia_rate'] +
    0.10 * df['women_anemia_rate'] +
    0.10 * df['child_diarrhea_rate'] +
    0.10 * df['child_ari_rate']
)
```

```python
df['basic_infrastructure_risk_score'] = (
    (1 - df['safe_drinking_water_access']) +
    (1 - df['sanitation_access']) +
    (1 - df['clean_cooking_fuel_access'])
) / 3
```

```python
df['lifestyle_disease_risk_score'] = (
    df['tobacco_use_men'] +
    df['tobacco_use_women'] +
    df['alcohol_use_men'] +
    df['alcohol_use_women']
) / 4
```

NCD Risk (Chronic Disease Risk)

```python
df['chronic_disease_risk_score'] = (
    df['high_blood_sugar_men'] +
    df['high_blood_sugar_women'] +
    df['high_blood_pressure_men'] +
    df['high_blood_pressure_women']
) / 4
```

```python
df['healthcare_accessibility_score'] = (
    df['institutional_delivery_rate'] +
    df['full_immunization_rate'] +
    df['health_insurance_coverage']
) / 3
```

```python
df['district_overall_health_risk_score'] = (
    0.4 * df['overall_child_health_risk_score'] +
    0.2 * df['basic_infrastructure_risk_score'] +
    0.2 * df['lifestyle_disease_risk_score'] +
    0.2 * df['chronic_disease_risk_score']
)
```

```python
def district_risk_level(score):
    if score < 0.3:
        return "Low"
    elif score < 0.6:
        return "Medium"
    else:
        return "High"

df['district_risk_level'] = df['district_overall_health_risk_score'].apply(district_risk_level)
```

```python
df['district_health_risk_rank'] = df['district_overall_health_risk_score'].rank(ascending=False)
```

```python
def identified_health_risk_factors(row):
    causes = []

    if row['safe_drinking_water_access'] < 0.5:
        causes.append("Poor drinking water")

    if row['sanitation_access'] < 0.5:
        causes.append("Poor sanitation")

    if row['child_stunting_rate'] > 0.4:
        causes.append("High malnutrition")

    if row['child_anemia_rate'] > 0.4:
        causes.append("High anemia")

    if row['clean_cooking_fuel_access'] < 0.5:
        causes.append("Indoor air pollution")

    if row['tobacco_use_men'] > 0.3:
        causes.append("High tobacco usage")

    return causes

df['identified_health_risk_factors'] = df.apply(identified_health_risk_factors, axis=1)
```
